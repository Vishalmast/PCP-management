```python
# Function to find all the specialties and their links to next page
def find_specialty():
    driver.get("https://www.vitals.com/")
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    all_specialty = []
    
    specialty = soup.find_all('a', attrs = {'class':'desktop-only'})
    p = str(specialty)
    index_of_specialty = [i for i in range(len(p)) if p.startswith('''data-browselist="specialty"''', i)]
    
    index_of_end_bracket = []
    
    for i in index_of_specialty:
        index = p[i:].find(">")
        index_of_end_bracket.append(index + i)
    index_of_start_bracket = []
    
    for i in index_of_end_bracket:
        index = p[i:].find("<")
        index_of_start_bracket.append(index + i)
        
    for i in range(len(index_of_end_bracket)):
        temp = p[index_of_end_bracket[i]+2: index_of_start_bracket[i]].strip()
        all_specialty.append(temp)
        
    specialty_link = []
    index_of_href = []
    
    for i in index_of_specialty:
        index = p[i:].find("href")
        index_of_href.append(index + i)
    index_of_start = []
    
    for  i in index_of_href:
        index = p[i:].find('''"''')
        index_of_start.append(index + i)
    index_of_end = []
    
    for  i in index_of_start:
        index = p[i+1:].find('''"''')
        index_of_end.append(index + i)
    for i in range(len(index_of_end)):
        temp = p[index_of_start[i]+1: index_of_end[i]+1].strip()
        specialty_link.append(temp)
    driver.quit()
    return all_specialty, specialty_link
```

```python
specialties, specialty_link = find_specialty()
```

```python
print(specialties, specialty_link, sep = "\n\n")
```

```python
for ind1 in range(1, len(specialties)):
    driver = webdriver.Chrome(path)
    path2 = domain_name + specialty_link[ind1]
    
    print(path2)
    driver.get(path2)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    states = []
    state_links = []
    
    state = soup.find_all('div', attrs = {'class':'card-panel'})
    p = str(state)
#     print(p)
    index_of_href = [i for i in range(len(p)) if p.startswith('''href''', i)]
    
    index_of_end = []
    for i in index_of_href:
        index = p[i:].find('''">''')
        index_of_end.append(index + i)
    index_of_end_a = []
    
    for i in index_of_href:
        index = p[i:].find("</a>")
        index_of_end_a.append(index + i)
        
    for i in range(len(index_of_end_a)):
        temp1 = p[index_of_href[i]+6: index_of_end[i]].strip()
        temp2 = p[index_of_end[i]+2 : index_of_end_a[i]].strip()
        states.append(temp2)
        state_links.append(temp1)
    
    index_of_popular_cities = []
    for index, i in enumerate(state_links):
        if i.find('''locations''') == -1:
            index_of_popular_cities.append(index)
    temp_state = states[:]
    temp_state_links = state_links[:]
    states = []
    state_links = []
    for index, ele in enumerate(temp_state_links):
        if index in index_of_popular_cities:
            continue
        else:
            state_links.append(ele)
    for index, ele in enumerate(temp_state):
        if index in index_of_popular_cities:
            continue
        else:
            states.append(ele)
            
#     print(states, state_links)
    driver.quit()
    for ind2 in range(len(states)):
        driver = webdriver.Chrome(path)
        path2 = domain_name + state_links[ind2]
        driver.get(path2)
        soup2 = BeautifulSoup(driver.page_source, 'html.parser')
        cities = []
        city_links = []
        city = soup2.find_all('div', attrs = {'class':'card-panel'})
        p = str(city)
        index_of_href = [i for i in range(len(p)) if p.startswith('''href''', i)]
        
        index_of_end = []
        for i in index_of_href:
            index = p[i:].find('''">''')
            index_of_end.append(index + i)
    
        index_of_end_a = []
        for i in index_of_href:
            index = p[i:].find("</a>")
            index_of_end_a.append(index + i)
        
        for i in range(len(index_of_end_a)):
            temp1 = p[index_of_href[i]+6: index_of_end[i]].strip()
            temp2 = p[index_of_end[i]+2 : index_of_end_a[i]].strip()
            cities.append(temp2)
            city_links.append(temp1)
#         print(cities, city_links)
        driver.quit()
        
        for ind3 in range(len(cities)):
            try:
                driver = webdriver.Chrome(path)
                path2 = domain_name + city_links[ind3]
                driver.get(path2)
                soup3 = BeautifulSoup(driver.page_source, 'html.parser')
            
                profile_links_res = []
                profile_links = soup3.find_all('a', attrs = {'class':''})
                p = str(profile_links)
                res = [i for i in range(len(p)) if p.startswith('''data-label="view_profile" data-non-webmd="view_profile"''', i)] 
                index_of_href = []
                for i in res:
                    index = p[i:].find("href")
                    index_of_href.append(index+i)
            
                index_of_view = []
                for i in index_of_href:
                    index = p[i:].find("View")
                    index_of_view.append(index + i)
                
                for i in range(len(index_of_view)):
                    temp = p[index_of_href[i] + 6: index_of_view[i] - 2]
                    temp = "www.vitals.com" + temp
                    profile_links_res.append(temp)
                
                doctor_names_res = []
                dr_names = soup3.find_all( attrs = {'class':'name'})
                p = str(dr_names)
                index_of_Dr = [i for i in range(len(p)) if p.startswith('''Dr''', i)]
            
                index_of_comma = []
                for i in index_of_Dr:
                    index = p[i:].find(",")
                    index_of_comma.append(index + i)
                
                for i in range(len(index_of_view)):
                    temp = p[index_of_Dr[i]: index_of_comma[i]]
                    doctor_names_res.append(temp)
                
                doctor_specialty_res = []
                specialty = soup3.find_all( attrs = {'class':'specialty uppercase'})
                p = str(specialty)
                index_of_end_bracket = [i for i in range(len(p)) if p.startswith('''>''', i)]
            
                index_of_start_bracket = []
                for i in index_of_end_bracket:
                    index = p[i:].find("<")
                    index_of_start_bracket.append(index + i)
                
                flag = 0
                for i in range(len(index_of_view)*2):
                    temp = p[index_of_end_bracket[i]+2: index_of_start_bracket[i]].strip()
                    if flag == 0:
                        doctor_specialty_res.append(temp)
                        flag = 1
                    else:
                        flag = 0
            
                ratings = []
                number_of_ratings = []
            
                rating = soup3.find_all('div', attrs = {'class':'card-rating'})
                for i in rating:
                    if len(ratings) == len(doctor_specialty_res):
                        break

                    if str(i).find("card-unrated") != -1:
                        ratings.append(None)
                        number_of_ratings.append(0)
                    else:
                        index = str(i).find('''class="overall"''')
                        index += 16
                        index2 = str(i)[index:].find('''</span>''')
                        index2 = index2 + index
                        ratings.append(float(str(i)[index:index2]))
    
                        index = str(i).find('''data-non-webmd="rating"''')
                        index2 = str(i)[index:].find('''</a>''')
                        index2 += index
        
                        index3 = str(i)[index:].find('''">''')
                        index3 += index
                        number_of_ratings.append(int(str(i)[index3 + 3 : index2-1]))
                    
                driver.quit()
                for ind4 in range(len(doctor_specialty_res)):
                    d = []
                    d.append(doctor_specialty_res[ind4])
                    d.append(doctor_names_res[ind4])
                    d.append(ratings[ind4])
                    d.append(number_of_ratings[ind4])
                    d.append(cities[ind3])
                    d.append(states[ind2])
                    print(d)
                    final_res.append(d)
            except:
                driver.quit()
                print("Error")
                continue
```
