```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

## Finding specialties

```python
all_specialty = []
```

```python
specialty = soup.find_all('a', attrs = {'class':'desktop-only'})
specialty
```

```python
p = str(specialty)
index_of_specialty = [i for i in range(len(p)) if p.startswith('''data-browselist="specialty"''', i)]
```

```python
index_of_end_bracket = []
for i in index_of_specialty:
    index = p[i:].find(">")
    index_of_end_bracket.append(index + i)
```

```python
index_of_start_bracket = []
for i in index_of_end_bracket:
    index = p[i:].find("<")
    index_of_start_bracket.append(index + i)
```

```python
for i in range(len(index_of_end_bracket)):
    temp = p[index_of_end_bracket[i]+2: index_of_start_bracket[i]].strip()
    all_specialty.append(temp)
```

```python
all_specialty
```

## Finding all links of specialty

```python
specialty_link = []
```

```python
index_of_href = []
for i in index_of_specialty:
    index = p[i:].find("href")
    index_of_href.append(index + i)
```

```python
index_of_start = []
for  i in index_of_href:
    index = p[i:].find('''"''')
    index_of_start.append(index + i)
```

```python
index_of_end = []
for  i in index_of_start:
    index = p[i+1:].find('''"''')
    index_of_end.append(index + i)
```

```python
print(index_of_start, index_of_end)
```

```python
for i in range(len(index_of_end)):
    temp = p[index_of_start[i]+1: index_of_end[i]+1].strip()
    specialty_link.append(temp)
```
